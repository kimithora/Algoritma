Encapsaulation

```python
class mahasiswa:
    def __init__(self, nama, nim, ipk):
        self.nama = nama #public
        self._nim = nim #protected
        self.__ipk = ipk #private

    def set_ipk(self, ipk):
        if 0 <= ipk <= 4:
            self.__ipk = ipk

    def get_ipk(self):
        return self.__ipk #reverse engineering

mhs1 = mahasiswa("kimi", 184250012, 3.5)
print(mhs1.nama) # Output: Andi
print(mhs1._nim) # Output: 184250012
print(mhs1.get_ipk) # Output: <bound method mahasiswa.get_ipk of <__main__.mahasiswa object at 0x...>>
print(mhs1.get_ipk()) # Output: 3.5
```

Abstraction

```python
from abc import ABC, abstractmethod

#abstract class
class transaksi(ABC):
    @abstractmethod
    def proses_transaksi(self):
        pass

#transfer
class transfer(transaksi):
    def __init__(self,rekening,jumlah):
        self.rekening = rekening
        self.jumlah = jumlah

    def proses_transaksi(self):
        print(f"Transfer sebesar {self.jumlah} ke rekening {self.rekening} telah diproses.")

#qris
class qris(transaksi):
    def __init__(self, merchant, jumlah):
        self.merchant = merchant
        self.jumlah = jumlah

    def proses_transaksi(self):
        print(f"QRIS ke {self.merchant} sebesar {self.jumlah} telah diproses.")

t1 = transfer("1234567890", 500000)
t2 = transfer("0987654321", 250000)

q1 = qris("Toko ABC", 150000)
q2 = qris("Restoran XYZ", 200000)

for transaksi in [t1, t2]:
    transaksi.proses_transaksi()

for transaksi in [q1, q2]:
    transaksi.proses_transaksi()
```
