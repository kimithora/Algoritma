```python
class mahasiswa:
    def __init__(self, nama, nim, jurusan):
        self.nama = nama
        self.nim = nim
        self.jurusan = jurusan

    def info(self):
        print(f"Nama: {self.nama}")
        print(f"NIM: {self.nim}")
        print(f"Jurusan: {self.jurusan}")

mahasiswa1 = mahasiswa("kimi", "184250012", "Sains Data")
mahasiswa1.info()    
```

```python
class penjualan:
    def __init__ (self, nama_produk, harga, jumlah, pajak):
        self.nama_produk = nama_produk
        self.harga = harga
        self.jumlah = jumlah
        self.pajak = pajak

    def belanjaa (self):
        total_harga = self.harga * self.jumlah
        total_pajak = total_harga * self.pajak
        total_belanja = total_harga + total_pajak
        print(f"produk yang dibeli adalah {self.nama_produk} dengan harga {self.harga} dan jumlah {self.jumlah}")
        print(f"total belanja yang harus dibayar adalah {total_belanja}")
        print(f"dikenakan pajak sebesar {total_pajak}")

keranjang1 = penjualan("saos ABC", 2500, 4, 0.11)
keranjang2 = penjualan("mie instan", 3500, 2, 0.11)
keranjang3 = penjualan("minyak goreng", 27000, 1, 0.11)
keranjang1.belanjaa()
keranjang2.belanjaa()
keranjang3.belanjaa()
```
